## 1️⃣ Variables & Basic Data Types

```python
a = 10          # Integer
b = 10.5        # Float
c = "Data"      # String
d = True        # Boolean

print(type(a), type(b), type(c), type(d))
```

## 2️⃣ Lists (Important for Pandas Series)

```python
marks = [70, 80, 90, 60]
print(marks)
```

```python
print(marks[0])
print(marks[-1])
```

```python
for m in marks:
    print(m)
```

```python
for m in marks:
    if m > 75:
        print("Above 75:", m)
```

```python
print("Sum:", sum(marks))
print("Length:", len(marks))
print("Max:", max(marks))
print("Min:", min(marks))
```

## 3️⃣ Dictionaries (Foundation of DataFrame)

```python
student_data = {
    "Name": ["Arun", "Bala", "Cathy"],
    "Age": [20, 21, 19],
    "Marks": [85, 78, 92]
}

print(student_data)
```

## 4️⃣ Conditional Statements (Used for Filtering)

```python
marks = 85

if marks > 50:
    print("Pass")
else:
    print("Fail")
```

```python
numbers = [10, 25, 40, 55]

for n in numbers:
    if n > 30:
        print(n)
```

## 5️⃣ Functions (Used in apply())

```python
def grade(mark):
    if mark >= 80:
        return "A"
    elif mark >= 60:
        return "B"
    else:
        return "C"

print(grade(85))
```

## 7️⃣ Mini Practice Exercises

```python
# Count numbers above 50
nums = [20, 55, 60, 45, 90]

count = 0
for n in nums:
    if n > 50:
        count += 1

print("Count:", count)
```

```python
# Find average
values = [10, 20, 30, 40]

average = sum(values) / len(values)
print("Average:", average)
```
