```python
df['Embarked'].mode()
```


# 📊 Step 19: Introduction to groupby()

groupby() allows us to:

Example: Survival rate by gender

Example: Survival rate by passenger class


# 📊 Step 20: Introduction to Matplotlib

We will create very simple plots.


## Plot 1: Survival Count

```python

survival_counts = df['Survived'].value_counts()

plt.figure()
plt.bar(['Did Not Survive','Survived'], survival_counts.values)
plt.title("Survival Count")
plt.ylabel("Number of Passengers")
plt.show()
```

## Plot 2: Age Distribution

```python

plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Plot 3: Survival Rate by Class

```python

class_survival = df.groupby('Pclass')['Survived'].mean()

plt.figure()
plt.bar(class_survival.index.astype(str), class_survival.values)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```
